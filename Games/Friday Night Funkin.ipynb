```python
import pygame
import random

# Initialisation de Pygame
pygame.init()

# Définition des constantes
WIDTH, HEIGHT = 500, 600
FPS = 60
WHITE = (255, 255, 255)
BLUE = (0, 0, 255)

# Initialisation de la fenêtre
win = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Friday Night Funkin'")

# Variables du jeu
notes = []
speed = 5
score = 0
note_interval = 50
note_speed = 10
note_delay = 0

# Boucle principale du jeu
def main():
    global note_delay, score
    clock = pygame.time.Clock()
    running = True

    while running:
        clock.tick(FPS)
        win.fill(WHITE)
        
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                running = False

        # Contrôle de la vitesse des notes
        note_delay += 1
        if note_delay >= note_interval:
            notes.append([random.randint(0, WIDTH - 100), -50])
            note_delay = 0

        # Mouvement des notes
        for note in notes:
            note[1] += note_speed

        # Nettoyage des notes sorties de l'écran
        for note in notes:
            if note[1] > HEIGHT:
                notes.remove(note)
                score += 1

        # Affichage du score
        font = pygame.font.SysFont(None, 30)
        text = font.render(f"Score: {score}", True, BLUE)
        win.blit(text, (10, 10))

        # Dessin des notes
        for note in notes:
            pygame.draw.rect(win, BLUE, (note[0], note[1], 100, 100))

        pygame.display.update()

    pygame.quit()

if __name__ == "__main__":
    main()
```
