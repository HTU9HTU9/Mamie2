```python
import pygame
import sys
import math
import pyaudio
import numpy as np

# Initialisation de Pygame
pygame.init()

# Définition des constantes
WIDTH, HEIGHT = 800, 600
DONUT_RADIUS = 200
LINE_WIDTH = 5
INNER_LINE_WIDTH = 10
RED = (255, 0, 0)
WHITE = (255, 255, 255)
BLACK = (0, 0, 0)
MOVEMENT_SPEED = 1
MICRO_THRESHOLD = 1000  # Seuil pour détecter le son du microphone

# Configuration de l'interface graphique
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Donut Game")

# Initialisation de l'audio
p = pyaudio.PyAudio()
FORMAT = pyaudio.paInt16
CHANNELS = 1
RATE = 44100
CHUNK = 1024

stream = p.open(format=FORMAT,
                channels=CHANNELS,
                rate=RATE,
                input=True,
                frames_per_buffer=CHUNK)

# Fonction pour dessiner le donut
def draw_donut():
    pygame.draw.circle(screen, WHITE, (WIDTH // 2, HEIGHT // 2), DONUT_RADIUS, LINE_WIDTH)

# Fonction pour dessiner les lignes tournantes
def draw_lines(angle_red, angle_white):
    red_x = WIDTH // 2 + int(DONUT_RADIUS * math.cos(math.radians(angle_red)))
    red_y = HEIGHT // 2 + int(DONUT_RADIUS * math.sin(math.radians(angle_red)))
    white_x = WIDTH // 2 + int(DONUT_RADIUS * math.cos(math.radians(angle_white)))
    white_y = HEIGHT // 2 + int(DONUT_RADIUS * math.sin(math.radians(angle_white)))
    
    # Dessiner les lignes intérieures avec une épaisseur augmentée
    pygame.draw.line(screen, RED, (WIDTH // 2, HEIGHT // 2), (red_x, red_y), INNER_LINE_WIDTH)
    pygame.draw.line(screen, WHITE, (WIDTH // 2, HEIGHT // 2), (white_x, white_y), INNER_LINE_WIDTH)

# Fonction principale du jeu
def main():
    angle_red = 0
    angle_white = 0

    clock = pygame.time.Clock()

    running = True
    while running:
        screen.fill(BLACK)

        # Gestion des événements
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                running = False

        # Déplacement automatique de la ligne rouge
        angle_red += 1

        # Capture audio
        data = np.frombuffer(stream.read(CHUNK), dtype=np.int16)
        volume = np.abs(data).mean()

        # Déplacement de la ligne blanche en fonction du volume du son
        if volume > MICRO_THRESHOLD:
            angle_white += MOVEMENT_SPEED
        
        # Mise à jour des angles de rotation
        angle_white %= 360
        angle_red %= 360

        # Dessin du jeu
        draw_donut()
        draw_lines(angle_red, angle_white)

        # Contrôle de la ligne blanche avec les touches directionnelles
        keys = pygame.key.get_pressed()
        if keys[pygame.K_LEFT]:
            angle_white -= MOVEMENT_SPEED
        elif keys[pygame.K_RIGHT]:
            angle_white += MOVEMENT_SPEED

        pygame.display.flip()

        clock.tick(60)

    pygame.quit()
    sys.exit()

if __name__ == "__main__":
    main()
```

```python
import pygame
import sys
import math
import pyaudio
import numpy as np
import winsound  # Pour les effets sonores

# Initialisation de Pygame
pygame.init()

# Définition des constantes
WIDTH, HEIGHT = 800, 600
DONUT_RADIUS = 200
LINE_WIDTH = 5
INNER_LINE_WIDTH = 10
RED = (255, 0, 0)
WHITE = (255, 255, 255)
BLACK = (0, 0, 0)
MOVEMENT_SPEED = 1
MICRO_THRESHOLD = 100  # Suggestion : essaie différentes valeurs pour ajuster le seuil
COUNTDOWN_SECONDS = 60  # Durée du chronomètre en secondes

# Configuration de l'interface graphique
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Donut Game")

# Initialisation de l'audio
p = pyaudio.PyAudio()
FORMAT = pyaudio.paInt16
CHANNELS = 1
RATE = 44100
CHUNK = 1024

stream = p.open(format=FORMAT,
                channels=CHANNELS,
                rate=RATE,
                input=True,
                frames_per_buffer=CHUNK)

# Fonction pour dessiner le donut
def draw_donut():
    pygame.draw.circle(screen, WHITE, (WIDTH // 2, HEIGHT // 2), DONUT_RADIUS, LINE_WIDTH)

# Fonction pour dessiner les lignes tournantes
def draw_lines(angle_red, angle_white):
    red_x = WIDTH // 2 + int(DONUT_RADIUS * math.cos(math.radians(angle_red)))
    red_y = HEIGHT // 2 + int(DONUT_RADIUS * math.sin(math.radians(angle_red)))
    white_x = WIDTH // 2 + int(DONUT_RADIUS * math.cos(math.radians(angle_white)))
    white_y = HEIGHT // 2 + int(DONUT_RADIUS * math.sin(math.radians(angle_white)))
    
    # Dessiner les lignes intérieures avec une épaisseur augmentée
    pygame.draw.line(screen, RED, (WIDTH // 2, HEIGHT // 2), (red_x, red_y), INNER_LINE_WIDTH)
    pygame.draw.line(screen, WHITE, (WIDTH // 2, HEIGHT // 2), (white_x, white_y), INNER_LINE_WIDTH)

# Fonction pour dessiner le chronomètre
def draw_timer(timer):
    font = pygame.font.SysFont(None, 30)
    text = font.render(f"Time left: {timer} seconds", True, RED)
    screen.blit(text, (10, 10))

# Fonction principale du jeu
def main():
    angle_red = 270  # Démarrer à l'opposé de la ligne blanche (à 9 heures sur une horloge)
    angle_white = 90
    timer = COUNTDOWN_SECONDS

    clock = pygame.time.Clock()

    running = True
    while running:
        screen.fill(BLACK)

        # Gestion des événements
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                running = False

        # Déplacement automatique de la ligne rouge
        angle_red += 1

        # Capture audio
        data = np.frombuffer(stream.read(CHUNK), dtype=np.int16)
        volume = np.abs(data).mean()

        # Déplacement de la ligne blanche en fonction du volume du son
        if volume > MICRO_THRESHOLD:
            angle_white += MOVEMENT_SPEED
        
        # Mise à jour des angles de rotation
        angle_white %= 360
        angle_red %= 360

        # Dessin du jeu
        draw_donut()
        draw_lines(angle_red, angle_white)
        draw_timer(timer)

        # Détection de collision entre la ligne rouge et la ligne blanche
        if abs(angle_red - angle_white) < 10:  # Ajuste la tolérance de collision au besoin
            winsound.Beep(500, 200)  # Jouer le son de bip

        pygame.display.flip()

        clock.tick(60)

        # Mise à jour du chronomètre
        timer -= 1
        if timer <= 0:
            running = False

    pygame.quit()
    sys.exit()

if __name__ == "__main__":
    main()
```
