```python
import random

def main():
    print("Bienvenue dans le jeu de pierre-papier-ciseaux !")
    print("Veuillez choisir :")
    print("1. Pierre")
    print("2. Papier")
    print("3. Ciseaux")

    choices = ["Pierre", "Papier", "Ciseaux"]

    while True:
        player_choice = input("Votre choix (1/2/3) : ")
        if player_choice not in ['1', '2', '3']:
            print("Veuillez entrer un choix valide (1, 2 ou 3) !")
            continue
        
        player_choice = int(player_choice) - 1
        computer_choice = random.randint(0, 2)

        print("\nVous avez choisi :", choices[player_choice])
        print("L'ordinateur a choisi :", choices[computer_choice])

        result = determine_winner(player_choice, computer_choice)
        if result == 0:
            print("Égalité !")
        elif result == 1:
            print("Vous avez gagné !")
        else:
            print("L'ordinateur a gagné !")

        play_again = input("Voulez-vous jouer à nouveau ? (o/n) : ")
        if play_again.lower() != 'o':
            break

def determine_winner(player, computer):
    if player == computer:
        return 0  # Égalité
    elif (player == 0 and computer == 2) or (player == 1 and computer == 0) or (player == 2 and computer == 1):
        return 1  # Le joueur gagne
    else:
        return -1  # L'ordinateur gagne

if __name__ == "__main__":
    main()
```
