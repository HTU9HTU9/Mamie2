```python
import random

class JeuPpopgi:
    def __init__(self):
        self.formes = ['Triangle', 'Cercle', 'Carré', 'Parapluie', 'Étoile']
        self.temps_limite = 10

    def jouer(self):
        print("Bienvenue dans le jeu Ppopgi !")
        print("Vous avez 10 minutes pour extraire parfaitement une forme estampée d'un Dalgona.")
        forme_correcte = random.choice(self.formes)
        print("La forme que vous devez extraire est :", forme_correcte)
        print("Commencez maintenant !")

        forme_extraite = input("Entrez la forme que vous avez extraite : ")

        if forme_extraite == forme_correcte:
            print("Félicitations ! Vous avez réussi à extraire la forme correcte.")
        else:
            print("Désolé, vous n'avez pas réussi à extraire la bonne forme.")

# Exemple d'utilisation du jeu
if __name__ == "__main__":
    jeu = JeuPpopgi()
    jeu.jouer()
```

```python
import pygame
import sys

# Initialisation de pygame
pygame.init()

# Définition de quelques couleurs
BLANC = (255, 255, 255)
NOIR = (0, 0, 0)
ROUGE = (255, 0, 0)

# Définition des dimensions de la fenêtre
largeur_fenetre = 600
hauteur_fenetre = 600
fenetre = pygame.display.set_mode((largeur_fenetre, hauteur_fenetre))
pygame.display.set_caption("Jeu Ppopgi")

# Chargement de l'image du Dalgona
image_dalgona = pygame.image.load("C:/Users/htu9h/Downloads/dalgona.png.png")

def afficher_texte(texte, taille, x, y, couleur=NOIR):
    font = pygame.font.SysFont(None, taille)
    texte_surface = font.render(texte, True, couleur)
    fenetre.blit(texte_surface, (x, y))

def jeu_ppopgi():
    clock = pygame.time.Clock()
    temps_limite = 600  # 600 frames (10 secondes à 60fps)
    temps_restant = temps_limite
    forme_extraite = False

    while True:
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                pygame.quit()
                sys.exit()

        # Affichage de l'image du Dalgona
        fenetre.blit(image_dalgona, (0, 0))

        # Affichage du temps restant
        afficher_texte("Temps restant: {} s".format(temps_restant // 60), 30, 10, 10)

        # Si le temps est écoulé, terminer le jeu
        if temps_restant <= 0:
            afficher_texte("Temps écoulé !", 50, 150, 250, ROUGE)
            pygame.display.update()
            pygame.time.wait(90000)  # Attendre 90 secondes
            return

        # Actualisation de l'affichage
        pygame.display.update()
        clock.tick(60)  # Limiter la vitesse de rafraîchissement à 60 FPS
        temps_restant -= 1

# Exemple d'utilisation du jeu
if __name__ == "__main__":
    jeu_ppopgi()
```

```python
import tkinter as tk
import random

class WhitePathGame:
    def __init__(self, master):
        self.master = master
        master.title("Reproduce the Path Game")

        self.canvas = tk.Canvas(master, width=800, height=800, bg="black", highlightthickness=0)
        self.canvas.pack()

        self.canvas.bind("<Button-1>", self.start_drawing)
        self.canvas.bind("<B1-Motion>", self.draw_line)
        self.canvas.bind("<ButtonRelease-1>", self.stop_drawing)

        self.path_coords = None
        self.drawing = False

        self.generate_random_path()
        self.player_path = []

    def generate_random_path(self):
        start_x = random.randint(50, 750)
        start_y = random.randint(50, 750)
        self.path_coords = [(start_x, start_y)]
        for _ in range(20):  # Longueur du chemin aléatoire
            direction = random.choice(["up", "down", "left", "right"])
            if direction == "up":
                start_y -= 30
            elif direction == "down":
                start_y += 30
            elif direction == "left":
                start_x -= 30
            elif direction == "right":
                start_x += 30
            self.path_coords.append((start_x, start_y))

        # Dessiner le chemin aléatoire
        self.canvas.create_line(self.path_coords, fill="white", width=12, tags="random_path")

    def start_drawing(self, event):
        if self.canvas.find_withtag(tk.CURRENT) and "random_path" in self.canvas.gettags(tk.CURRENT):
            self.drawing = True
            self.current_line = [(event.x, event.y)]
            self.canvas.delete("user_path")

    def draw_line(self, event):
        if self.drawing:
            self.current_line.append((event.x, event.y))
            self.canvas.delete("user_path")
            self.canvas.create_line(self.current_line, fill="red", width=12, tags="user_path")

    def stop_drawing(self, event):
        if self.drawing:
            self.drawing = False
            self.canvas.delete("user_path")
            # Vérifier si le chemin de l'utilisateur correspond au chemin aléatoire
            if self.current_line == self.path_coords[:len(self.current_line)]:
                print("Partie du chemin correcte, continuez !")
                self.player_path.extend(self.current_line)
                if len(self.player_path) == len(self.path_coords):
                    print("Félicitations, vous avez reproduit le chemin !")
                    self.master.quit()
            else:
                print("Désolé, votre chemin ne correspond pas à cette partie.")
                self.player_path = []

def main():
    root = tk.Tk()
    game = WhitePathGame(root)
    root.mainloop()

if __name__ == "__main__":
    main()
```

```python
import tkinter as tk
import random

class WhitePathGame:
    def __init__(self, master):
        self.master = master
        master.title("Coloriage Game")

        self.canvas = tk.Canvas(master, width=800, height=800, bg="black", highlightthickness=0)
        self.canvas.pack()

        self.path_coords = []
        self.pixel_grid = []  # Liste de tous les pixels du chemin

        self.generate_random_path()

        self.canvas.bind("<B1-Motion>", self.color_pixel)

    def generate_random_path(self):
        start_x = random.randint(50, 750)
        start_y = random.randint(50, 750)
        self.path_coords.append((start_x, start_y))
        for _ in range(15):  # Longueur du chemin aléatoire
            direction = random.choice(["up", "down", "left", "right"])
            if direction == "up":
                start_y -= 40
            elif direction == "down":
                start_y += 40
            elif direction == "left":
                start_x -= 40
            elif direction == "right":
                start_x += 40
            self.path_coords.append((start_x, start_y))

        # Créer une grille de pixels pour chaque segment de ligne
        for i in range(len(self.path_coords) - 1):
            x1, y1 = self.path_coords[i]
            x2, y2 = self.path_coords[i+1]
            for j in range(21):  # Diviser chaque segment en 20 petits pixels
                x = x1 + (x2 - x1) * j / 20
                y = y1 + (y2 - y1) * j / 20
                self.pixel_grid.append((x, y))

        # Dessiner le chemin aléatoire
        for i in range(len(self.path_coords) - 1):
            self.canvas.create_line(self.path_coords[i], self.path_coords[i+1], fill="white", width=20, tags="path")

    def color_pixel(self, event):
        x, y = event.x, event.y
        item = self.canvas.find_closest(x, y)[0]
        item_tags = self.canvas.gettags(item)
        if "path" in item_tags:
            if self.canvas.itemcget(item, "fill") != "red":
                self.canvas.itemconfig(item, fill="red")
                self.pixel_grid.remove((x, y))
                self.check_completion()

    def check_completion(self):
        if not self.pixel_grid:  # Si la liste des pixels est vide, le chemin est entièrement rouge
            print("Félicitations, vous avez colorié tout le chemin en rouge !")
            self.master.quit()

def main():
    root = tk.Tk()
    game = WhitePathGame(root)
    root.mainloop()

if __name__ == "__main__":
    main()
```
