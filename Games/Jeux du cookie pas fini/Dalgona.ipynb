```python
from PIL import Image, ImageDraw
import math

# Créer une nouvelle image avec un fond noir
largeur, hauteur = 900, 900
image_hexagone = Image.new("RGB", (largeur, hauteur), "black")

# Dessiner un hexagone blanc avec un contour noir centré sur l'image
longueur_cote = 500
rayon = longueur_cote / math.sqrt(3)
centre_x, centre_y = largeur // 2, hauteur // 2
draw = ImageDraw.Draw(image_hexagone)
epaisseur_contour = 20  # Changer cette valeur pour ajuster l'épaisseur du contour

# Coordonnées des sommets de l'hexagone
sommets = []
for i in range(6):
    angle = 2 * math.pi / 6 * i
    sommets.append((centre_x + rayon * math.cos(angle), centre_y + rayon * math.sin(angle)))

# Tracer les côtés de l'hexagone avec l'épaisseur du contour
for i in range(6):
    draw.line([sommets[i], sommets[(i + 1) % 6]], fill="white", width=epaisseur_contour)

# Relier le dernier côté au premier côté pour fermer la forme
draw.line([sommets[5], sommets[0]], fill="white", width=0)

# Enregistrer l'image
image_hexagone.save("hexagone.png")
```

```python
import pygame
import os
import random

# Initialisation de Pygame
pygame.init()

# Définir les dimensions de la fenêtre
largeur, hauteur = 800, 600
fenetre = pygame.display.set_mode((largeur, hauteur))
pygame.display.set_caption("Jeu du Dalgona")

# Couleurs
BLANC = (255, 255, 255)
ROUGE = (255, 0, 0)

# Fonction principale du jeu
def jouer_jeu():
    # Spécifier le chemin complet vers le dossier contenant les images géométriques
    chemin_dossier_geometrique = "C:\\Users\\htu9h\\Downloads\\Formes géométrique"

    # Charger la liste des chemins d'accès des images géométriques dans le dossier
    chemins_images = [os.path.join(chemin_dossier_geometrique, fichier) for fichier in os.listdir(chemin_dossier_geometrique) if fichier.endswith(".png")]

    # Vérifier si des images ont été trouvées
    if chemins_images:
        # Choisir une image au hasard parmi la liste des chemins
        chemin_image_aleatoire = random.choice(chemins_images)

        # Charger l'image choisie au hasard
        image_aleatoire = pygame.image.load(chemin_image_aleatoire).convert_alpha()

        # Afficher l'image au centre de la fenêtre
        rect = image_aleatoire.get_rect(center=(largeur // 2, hauteur // 2))
        fenetre.fill(BLANC)
        fenetre.blit(image_aleatoire, rect)
        pygame.display.flip()

        # Variable pour le traçage continu
        traçage_en_cours = False

        # Attendre jusqu'à ce que le joueur ferme la fenêtre
        en_cours = True
        while en_cours:
            for event in pygame.event.get():
                if event.type == pygame.QUIT:
                    en_cours = False

                # Commencer le traçage si le bouton gauche de la souris est enfoncé
                if event.type == pygame.MOUSEBUTTONDOWN and event.button == 1:
                    position_souris = pygame.mouse.get_pos()
                    couleur_pixel = image_aleatoire.get_at(position_souris)
                    # Vérifier si la couleur du pixel est noire (transparent)
                    if rect.collidepoint(position_souris) and couleur_pixel.a == 0:
                        traçage_en_cours = True

                # Continuer le traçage si le bouton gauche de la souris est maintenu enfoncé
                elif event.type == pygame.MOUSEMOTION and traçage_en_cours:
                    position_souris = pygame.mouse.get_pos()
                    couleur_pixel = image_aleatoire.get_at(position_souris)
                    # Vérifier si la couleur du pixel est noire (transparent)
                    if rect.collidepoint(position_souris) and couleur_pixel.a == 0:
                        pygame.draw.circle(fenetre, ROUGE, position_souris, 5)
                        pygame.display.flip()

                # Arrêter le traçage lorsque le bouton gauche de la souris est relâché
                elif event.type == pygame.MOUSEBUTTONUP and event.button == 1:
                    traçage_en_cours = False

    else:
        print("Aucune image trouvée dans le dossier spécifié.")

# Jouer le jeu
jouer_jeu()

# Quand le joueur ferme la fenêtre, quitter Pygame
pygame.quit()
```
